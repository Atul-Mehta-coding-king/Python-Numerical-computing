```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
```

```python
def BallVertical(Vo):
  g = 9.81
  y = 0
  t = 0
  time = []
  location = []
  while (y>=0):
    # print('At t = ' , str(t),'Location ', y)
    t += 0.1
    y = (Vo*t) - ((g*(t**2))/2)
    time.append(t)
    location.append(y)
  plt.plot(time,location)
  plt.xlabel('time')
  plt.ylabel('Location')
  plt.title("Y vs T")
  plt.show
```

```python
BallVertical(20)

BallVertical(40)

BallVertical(60)
```
